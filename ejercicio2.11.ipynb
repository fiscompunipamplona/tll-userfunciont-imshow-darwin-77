```
def factorial(n):
    s=1.0
    for i in range(1,n+1):
        s *=i
    return s

def binomial(n,k):
    if k>0:
        res = factorial(n)/(factorial(k)*factorial(n-k))
    elif k == 0:
        res = 1
    elif k < 0:
        return(print("k<0; no lo calcula"))
    return(int(res))
```
