example 3.1 wave interference\

suppose we drop a pebble in a pond and waves radiate out from the spot where it fell we could create a simple representation of the physics with a sine wave spreading out in a uniform circle to represent the height of the waves at some later time. if the center of the circle is at x1,x2 then the distance r1 to the center from a point x,y is\
r=√(x−x1)2+ (y−y1)2\

and the wave for the height is\

c1(x, y) =c0sinkr1\

where c0 is the amplitude of the waves and k is the wavevector related to the wavelength h by k=2π/h\
now suppose we drop another pebble in the pond creating another circular set of waves with the same wavelength and amplitude but centered on a different point x2,y2\

r=√(x−x1)2+(y−y1)2)  with c1(x, y)=c0sinkr1\

then assuming the waves add linearly(which is a reasonable assumption for water waves provided they are not too big) the total heigth of the surface at a point x,y is\

c(x, y) =c0sinkr1+c0sinkr2\

suppose the wavelength of the waves is h=5cm the amplitude is 1cm and the centers of the circles are 20cm apart here is a program to make an image of the heigth over a 1m square region of the pond to make the image we create an array of values representing the heigth c at a grid of points and then use that array to make a density plot in this example we use a grid of 500x500 points to cover the 1m square which means the grid points have a separation of 100/500=0.2cm\

ejemplo 3.1 interferencia de onda\

supongamos que dejamos caer un guijarro en un estanque y las olas irradian desde el lugar donde cayó, podríamos crear una representación simple de la física con una onda sinusoidal que se extienda en un círculo uniforme para representar la altura de las olas en algún momento posterior. si el centro del círculo está en x1, x2, entonces la distancia r1 al centro desde un punto x, y es\

r=√(x−x1)2+ (y−y1)2)\

y la ola para la altura es\

c1(x, y) =c0sinkr1\

donde c0 es la amplitud de las ondas yk es el vector de onda relacionado con la longitud de onda h por k = 2π / h
ahora supongamos que dejamos caer otra piedra en el estanque creando otro conjunto circular de ondas con la misma longitud de onda y amplitud pero centradas en un punto diferente x2, y2\

c1(x, y) =c0sinkr1  con  r=√(x−x1)2+(y−y1)\

luego, suponiendo que las ondas se suman linealmente (lo cual es una suposición razonable para las ondas de agua siempre que no sean demasiado grandes), la altura total de la superficie en un punto x, y es\

c(x, y) =c0sinkr1+c0sinkr2\

supongamos que la longitud de onda de las ondas es h = 5 cm, la amplitud es de 1 cm y los centros de los círculos están separados por 20 cm aquí hay un programa para hacer una imagen de la altura sobre una región cuadrada de 1 m del estanque para hacer que la imagen creemos una matriz de valores que representan la altura c en una cuadrícula de puntos y luego usar esa matriz para hacer un diagrama de densidad en este ejemplo usamos una cuadrícula de 500x500 puntos para cubrir el cuadrado de 1 m, lo que significa que los puntos de la cuadrícula tienen una separación de 100/500 = 0.2 cm


```
#from math import sqrt, sin,pi
import math
import numpy 
#from pylab import imshow, gray, show,hot, figure
import matplotlib.pyplot as plt

wavelength=5.0
k=2*math.pi/wavelength
xi0=1
separation=20
side=100
points=500
spacing=side/points
#calculando la posisicion de los centros de los circulos

x1=side/2 + separation/2
y1=side/2
x2=side/2 - separation/2
y2=side/2

#creando los vectores 
xi=numpy.empty([points,points],float)

#calculando los valores de los vectores 
for i in range (points):
    y=spacing*i
    for j in range (points):
        x=spacing*j
        r1=math.sqrt((x-x1)**2 + (y-y1)**2)
        r2=math.sqrt((x-x2)**2 + (y-y2)**2)
        xi[i,j]= xi0*math.sin(k*r1) + xi0*math.sin(k*r2)
    
#realizando el plot
f = plt.figure()
#plt.imshow(xi,origin="lower",extent=[0,side,0,side])
plt.imshow(xi,origin="lower",extent=[0,side,0,side])
#green()
#show()
plt.imsave('Phase.jpg', xi, cmap='viridis')
plt.imsave('Phase.jpg', xi, cmap='viridis')
```
