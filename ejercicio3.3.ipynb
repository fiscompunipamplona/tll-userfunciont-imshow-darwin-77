hay un archivo de recursos en línea llamado smt.txt que contiene una cuadrícula de valores de las mediciones del microscopio de túnel de exploración de la superficie (111) de silicio. Un microscopio de túnel de exploración (STM) es un dispositivo que mide la forma de una superficie en el nivel atómico al rastrear una punta afilada sobre la superficie y asegurar la corriente de túnel cuántico en función de la posición, el resultado final es una cuadrícula de valores que representan la altura de la superficie y el archivo smt.txt contiene exactamente esa cuadrícula de valores wirte a programa que lee los datos contenidos en el archivo y hace un diagrama de densidad de los valores, use las diversas opciones y variantes que ha aprendido para hacer una imagen que muestre claramente la estructura de la superficie del silicio


```
#from pylab import imshow, show, inferno,pink,summer
import matplotlib.pyplot as plt
from numpy import loadtxt


data = loadtxt("Datos_FC/datos.txt",float)
f = plt.figure()
plt.imshow(data, origin="lower")
plt.summer()
show()
plt.imsave('fig.jpg', data, cmap='viridis')
f = plt.figure()
plt.imshow(data, origin="lower", extent=[320,380,220,280])
plt.inferno()
show()
plt.imsave('Phase.jpg', data, cmap='viridis')
```
